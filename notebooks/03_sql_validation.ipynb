```python
# Imports & Connection
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine(
    "postgresql://postgres:Vivek123@localhost:5432/retail_dw"
)

print("Connected Successfully")
```

```python
# Orders Count Validation
query = """
SELECT COUNT(*) as total_orders
FROM raw.orders
"""

pd.read_sql(query, engine)
```

```python
# Revenue Validation
query = """
SELECT ROUND(SUM(payment_value),2) as total_revenue
FROM raw.payments
"""

pd.read_sql(query, engine)
```

```python
# State Revenue Validation
query = """
SELECT *
FROM analytics.fct_state_revenue
ORDER BY total_revenue DESC
LIMIT 10
"""

pd.read_sql(query, engine)
```

```python
# Monthly Revenue Validation
query = """
SELECT *
FROM analytics.fct_monthly_revenue
ORDER BY year, month
"""

pd.read_sql(query, engine)
```

```python
# Order Status Validation
query = """
SELECT *
FROM analytics.fct_order_status
ORDER BY total_orders DESC
"""

pd.read_sql(query, engine)
```
