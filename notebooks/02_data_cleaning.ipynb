```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv("../data/raw/kaggle/olist/olist_orders_dataset.csv")

customers = pd.read_csv("../data/raw/kaggle/olist/olist_customers_dataset.csv")

order_items = pd.read_csv("../data/raw/kaggle/olist/olist_order_items_dataset.csv")

products = pd.read_csv("../data/raw/kaggle/olist/olist_products_dataset.csv")

payments = pd.read_csv("../data/raw/kaggle/olist/olist_order_payments_dataset.csv")

print("All datasets loaded successfully")
```

```python
# Duplicate Check
print("Orders Duplicates:", orders.duplicated().sum())

print("Customers Duplicates:", customers.duplicated().sum())

print("Order Items Duplicates:", order_items.duplicated().sum())

print("Products Duplicates:", products.duplicated().sum())

print("Payments Duplicates:", payments.duplicated().sum())
```

```python
# Missing Values Check
print("Orders Missing Values")
print(orders.isnull().sum())

print("\nCustomers Missing Values")
print(customers.isnull().sum())

print("\nOrder Items Missing Values")
print(order_items.isnull().sum())

print("\nProducts Missing Values")
print(products.isnull().sum())

print("\nPayments Missing Values")
print(payments.isnull().sum())
```

```python
# Missing Values Percentage
def missing_percentage(df):

    missing = (df.isnull().sum() / len(df)) * 100

    return missing.round(2)

print(missing_percentage(products))
```

```python
# Datetime Conversion (Orders)
date_columns = [

    'order_purchase_timestamp',

    'order_approved_at',

    'order_delivered_carrier_date',

    'order_delivered_customer_date',

    'order_estimated_delivery_date'

]

for col in date_columns:

    orders[col] = pd.to_datetime(
        orders[col],
        errors='coerce'
    )

orders.info()
```

```python
# Products Missing Values Treatment
products['product_category_name'] = products[
    'product_category_name'
].fillna('Unknown')
```

```python
products['product_name_lenght'] = products[
    'product_name_lenght'
].fillna(
    products['product_name_lenght'].median()
)

products['product_description_lenght'] = products[
    'product_description_lenght'
].fillna(
    products['product_description_lenght'].median()
)

products['product_photos_qty'] = products[
    'product_photos_qty'
].fillna(
    products['product_photos_qty'].median()
)
```

```python
# Weight & Dimensions Missing Values
products['product_weight_g'] = products[
    'product_weight_g'
].fillna(
    products['product_weight_g'].median()
)

products['product_length_cm'] = products[
    'product_length_cm'
].fillna(
    products['product_length_cm'].median()
)

products['product_height_cm'] = products[
    'product_height_cm'
].fillna(
    products['product_height_cm'].median()
)

products['product_width_cm'] = products[
    'product_width_cm'
].fillna(
    products['product_width_cm'].median()
)
```

```python
# Validation
print(products.isnull().sum())
```

```python
# Remove Duplicates (Safe)
orders = orders.drop_duplicates()

customers = customers.drop_duplicates()

order_items = order_items.drop_duplicates()

products = products.drop_duplicates()

payments = payments.drop_duplicates()
```

```python
# Create Cleaned Folder Output
orders.to_csv(
    "../data/processed/cleaned/orders_cleaned.csv",
    index=False
)

customers.to_csv(
    "../data/processed/cleaned/customers_cleaned.csv",
    index=False
)

order_items.to_csv(
    "../data/processed/cleaned/order_items_cleaned.csv",
    index=False
)

products.to_csv(
    "../data/processed/cleaned/products_cleaned.csv",
    index=False
)

payments.to_csv(
    "../data/processed/cleaned/payments_cleaned.csv",
    index=False
)

print("Cleaned files exported successfully")
```
